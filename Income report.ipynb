```python
import pandas as pd
```

```python
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

```python
# Here i am displaying the first 5 observations from the DataFrame to see the data im working with
df.head()
```

```python
# Here i am displaying the first 20 values of the series for Ethnicity
# This is so that i can see what Ethnicities there are
df.Ethnicity.head(20)
```

```python
# Here I am printing the averages i have calculated for each Ethnicity
print("The average income of a Caucasion person is", df[df.Ethnicity == "Caucasian"].loc[:,"Income"].mean())
print("The average income of a Asian person is",df[df.Ethnicity == "Asian"].loc[:,"Income"].mean())
print("The average income of a African Amercian person is",df[df.Ethnicity == "African American"].loc[:,"Income"].mean())
```

```python
# Here i am 
print("The average balance of a Married person is", df[df.Married == "Yes"].loc[:,"Balance"].mean())
print("The average balance of a Single person is", df[df.Married == "No"].loc[:,"Balance"].mean())
print("This shows that the average balance of as Single person is higher than a Married person")
```

```python
df.sort_values(by='Income',ascending=False).head()

# The highest income is 186.63 - found in the 28th row index
```

```python
df.sort_values(by='Income',ascending=True).head()

# The lowest income is 10.354 - found in the 58th row index
```

```python
print("The total number of cards is", df.loc[:,"Cards"].sum())
```
